```python
from bs4 import BeautifulSoup
import requests
import re



callParticipantsList = []
preparedRemarksContentList = []
questionAnswerContentList = []
companyNameList = []
tickerList = []
dateList = []
timeList=[]
quarterList=[]

listofurls = []
newlistofurls = []

for x in range(1,3): #Takes time to load all 677 pages : so try only 3 pages here. fill free to change it
    url = 'https://www.fool.com/earnings-call-transcripts/?page='+str(x)
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page, "lxml")

    for a in soup.find_all("a", {'class': 'text-gray-1100', 'href': lambda href: href and href.startswith('/earnings/')}):
        listofurls.append('https://www.fool.com'+ a['href'])


for x in listofurls:
    
    url = str(x)
    print(url)
    response = requests.get(url)
    page = response.text
    soup = BeautifulSoup(page, "lxml")
    
    companyName = [soup.find(class_='tailwind-article-body').find_all('p')[1].find('strong').text]
    print(companyName)
    for x in companyName:
        companyNameList.append(x)

    ticker = [soup.find(class_='tailwind-article-body').find_all('p')[1].find(class_='ticker-symbol').text]
    for x in ticker:
        tickerList.append(x)
    # print(ticker)

    date = [soup.find(class_='tailwind-article-body').find(id='date').text]
    for x in date:
        dateList.append(x)
    # print(date)
    
    time = [soup.find(class_='tailwind-article-body').find(id='time').text]
    for x in time:
        timeList.append(x)
    # print(time)
    
    # quarter = [str(soup.find_all('p')[3])[str(soup.find_all('p')[3]).find("<br/>"):
    #                        str(soup.find_all('p')[3]).find('<span id="date">')].replace("<br/>","").replace("<br/>","")[0:2]]
    quarterBulkText = [soup.find(class_='tailwind-article-body').find_all('p')[1].get_text(strip=True)]
    pattern = r'Q[1-4]\s\d{4}\sEarnings Call'
    quarterText = re.search(pattern, quarterBulkText[0]).group(0)
    quarter = [quarterText.replace('\xa0', ' ')]
    for x in quarter:
        quarterList.append(x)
    # print(quarterList)

    namesList = []
    rolesList = []

    names = soup.select('.tailwind-article-body p:has(strong)')  # select all <p> tags with a "strong" tag inside
    names = [name for name in names if len(name.select_one('strong').text.split()) == 2]  # filter only those with 2 words in the "strong" tag

    # print(names)
    for name in names:
        text = name.text
        try:
            name, title = text.split(" -- ")

        except:
            try:
             name,company,title=text.split(" -- ")
            except:
                continue
        namesList.append(name)
        rolesList.append(title)
    # print(len(namesList))
    # print(len(rolesList))
    callParticipants = [list(zip(namesList, rolesList))]
    for x in callParticipants:
        callParticipantsList.append([x])
    toc = []
    for x in soup.find(class_='tailwind-article-body').find_all('h2'):
        toc.append(x.text)
    if 'Questions & Answers:' or 'Questions and Answers:'in toc:
        qanda = []
        for div in soup.find(class_='tailwind-article-body').find_all('h2')[2].find_all_next('p'):
            qanda.append(div.text)
        for x in qanda:
            if 'Duration:' in x:
                durationPlacement = qanda.index(x)
        QAContent = qanda[:durationPlacement] 
        speaker = []
        for x in range(0,len(QAContent),2):
            speaker.append(QAContent[x])
        qaContent = []
        for x in range(1,len(QAContent),2):
            qaContent.append(QAContent[x])
        questionAnswerContent = [list(zip(speaker,qaContent))]
        for x in questionAnswerContent:
                questionAnswerContentList.append(x)
    else:
        questionAnswerContentList.append(0)
    # print(questionAnswerContentList)
    preparedRemarks = []
    for div in soup.find(class_='tailwind-article-body').find_all('p')[4:]:
        preparedRemarks.append(div.text)
    prIndex = []
    for x in preparedRemarks:
        if 'Operator' in x:
            prIndex.append(preparedRemarks.index(x))
        else:
            None
    if len(prIndex) == 0:
        None
    else:
        preparedRemarksContent = [preparedRemarks[:prIndex[1]]]
    for x in preparedRemarksContent:
        preparedRemarksContentList.append(x)
```
